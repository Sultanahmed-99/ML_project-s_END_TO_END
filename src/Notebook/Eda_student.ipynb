###  3.8 (Feature Engineering) Adding columns for ""Total Score" and "Average"


```python
# adding Toatl score which could as y target to predict  
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
# adding average of total score and average score could be y as target to predict 
df['Average_score'] = df['total_score'] / 3 

df.sample(4)
```

##### Building more features to test does count of how many 
    student get 100 score over math_score , reading_score , writing_score
        effect the y target varible 

```python
#  count of students how get's average score in 
    # math , reading , writing equal to 100 full 
math_full = df.query('math_score == 100')['Average_score'].count()
reading_full = df.query('reading_score == 100')['Average_score'].count()
writing_full = df.query('writing_score == 100')['Average_score'].count()

print(f'Number of students with full mark in math : {math_full}')
print(f'Number of students with full mark in reading : {reading_full}')
print(f'Number of students with full mark in writing : {writing_full}')
```

```python
#  count of students how get's average score in 
    # math , reading , writing less or equal to 50 full 
math_less_20= df.query('math_score <= 20')['Average_score'].count()
reading_full_less_20 = df.query('reading_score <=  20')['Average_score'].count()
writing_full_less_20 = df.query('writing_score <=  20')['Average_score'].count()

print(f'Number of students with full mark in math : {math_less_20}')
print(f'Number of students with full mark in reading : {reading_full_less_20}')
print(f'Number of students with full mark in writing : {writing_full_less_20}')
```

## Insight 
- From above values we get students have performed the worst in Maths section 
- Best performanec is in reading section
